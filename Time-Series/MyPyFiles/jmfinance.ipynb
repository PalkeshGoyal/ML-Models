```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
jmfinance = pd.read_csv("../JMFINANCIL.NS.csv", header=0)
```

```python
jmfinance = jmfinance[['High' , 'Low']]
jmfinance.head()
```

```python
jmfinance_high = jmfinance[['High']]
jmfinance_low = jmfinance[['Low']]
```

```python
for i in range(1,8):
    high_name = "high_shift"+str(i)
    low_name = "low_shift"+str(i)
    jmfinance_high[high_name] = jmfinance['High'].shift(i)
    jmfinance_high[low_name] = jmfinance['Low'].shift(i)
    jmfinance_low[high_name] = jmfinance['High'].shift(i)
    jmfinance_low[low_name] = jmfinance['Low'].shift(i)
```

```python
jmfinance_high.dropna(inplace=True)
jmfinance_low.dropna(inplace=True)
```

```python
X_high , y_high = jmfinance_high.drop("High",axis=1) , jmfinance_high['High']
X_low , y_low = jmfinance_low.drop("Low",axis=1) , jmfinance_low['Low']
```

```python
from sklearn.linear_model import LinearRegression
model = {}
model['High'] = LinearRegression()
model['Low'] = LinearRegression()
```

```python
model['High'].fit(X_high , y_high)
model['Low'].fit(X_low , y_low)
```

```python
for days in range(10):
    new_row = {}
    for i in range(1,8):
        high_name = "high_shift"+str(i)
        low_name = "low_shift"+str(i)
        new_row[high_name] = [jmfinance['High'].iloc[-i]]
        new_row[low_name] = [jmfinance['Low'].iloc[-i]]
    new_df = pd.DataFrame.from_dict(new_row)
    df_new_row = {}
    df_new_row['High'] = [model['High'].predict(new_df)[0]]
    df_new_row['Low'] = [model['Low'].predict(new_df)[0]]
    df_new_row = pd.DataFrame.from_dict(df_new_row)
    jmfinance = pd.concat([jmfinance , df_new_row] , ignore_index= True )
```
